```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.html.widgets import *
from __future__ import division
from __future__ import print_function
```

```python
filename1 = 'BufferSolutionRayleighScattering.csv'
filename2 = 'BabyPowder.csv'
data1 = np.loadtxt(filename1, delimiter=',', skiprows=2)
data2 = np.loadtxt(filename2, delimiter=',', skiprows=2)
```

```python
data1[0,:]
```

```python
data1[:,0] = data1[:,0] - data1[0,0]
```

```python
data1[0:]
```

```python
data2[:,0] = data2[:,0] - data2[0,0]
```

```python
print(data1.shape,data1.shape[0])
```

```python
npts = data1.shape[0]
nstart = 0
npts_in_plot = 1e6
n1,n2 = nstart,nstart+npts_in_plot
fig, ax = plt.subplots(1,2,figsize=(14,6))
ax[0].plot(data1[n1:n2,0]/1000,data1[n1:n2,1])
ax[1].plot(data1[n1:n2,0]/1000,data1[n1:n2,2])
for a in ax:
    a.set_ylim(0,3.3)
    a.set_xlabel('Voltage (V)')
    a.set_xlabel('Time (ms)')
```

```python

npts = data1.shape[0]
nstart = 5.75e5
npts_in_plot = 5e2
def plotsubrange(nstart,npts_in_plot):
    n1,n2 = nstart,nstart+npts_in_plot
    fig, ax = plt.subplots(1,2,figsize=(14,6))
    ax[0].plot(data2[n1:n2,0]/1000,data2[n1:n2,1])
    ax[1].plot(data2[n1:n2,0]/1000,data2[n1:n2,2])
    for a in ax:
        a.set_ylim(0,3.3)
        a.set_xlabel('Voltage (V)')
        a.set_xlabel('Time (ms)')
interact(plotsubrange, nstart=(0,2.8e6,1e4), npts_in_plot=(100,2000,100));
```
