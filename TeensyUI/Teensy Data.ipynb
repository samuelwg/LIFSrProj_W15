```python
from pyqtgraph import QtGui, QtCore
import pyqtgraph as pg
import sys
import serial
from datetime import datetime

## Always start by initializing Qt (only once per application)
app = QtGui.QApplication([])

## Define a top-level widget to hold everything
w = QtGui.QWidget()
w.resize(1000,600)
w.setWindowTitle('Polarization Visualization')

resetBtnClicked = False
startBtnClicked = False

def startButtonClicked():
    global startBtnClicked
    global startBtn
    if (startBtnClicked == False):
        startBtnClicked = True
            
    elif (startBtnClicked == True):
        startBtnClicked = False
    
def resetButtonClicked():
    global resetBtnClicked
    resetBtnClicked = True
    
## Create widgets to be placed inside

startBtn = QtGui.QPushButton('Start')
startBtn.setToolTip('Click to begin graphing')
startBtn.resize(startBtn.sizeHint())
        
resetBtn = QtGui.QPushButton('Reset')
resetBtn.setToolTip('Click to clear graph')
resetBtn.resize(resetBtn.sizeHint())

startBtn.clicked.connect(startButtonClicked)
resetBtn.clicked.connect(resetButtonClicked)

# Create plot widget
wGL = pg.PlotWidget()
wGL.setSizePolicy(QtGui.QSizePolicy.Expanding, QtGui.QSizePolicy.Expanding)
wGL.setYRange(0, 3.3)
#wGL.setLabel('top', text = "Plot 1")
curve = wGL.plot()

wGL2 = pg.PlotWidget()
wGL2.setSizePolicy(QtGui.QSizePolicy.Expanding, QtGui.QSizePolicy.Expanding)
wGL2.setYRange(0, 3.3)
#wGL.setLabel('top', text = "Plot 2")
curve2 = wGL2.plot()

## Create a grid layout to manage the widgets size and position
layout = QtGui.QGridLayout()
w.setLayout(layout)
#layout.setColumnStretch (1, 2)

## Add widgets to the layout in their proper positions
layout.addWidget(startBtn, 1, 0)   # start button
layout.addWidget(resetBtn, 2, 0)   #reset button
layout.addWidget(wGL, 1, 1, 1, 1)  # wGL goes on right side, spanning 3 rows
layout.addWidget(wGL2, 1, 2, 1, 1)

## Display the widget as a new window
w.show()

teensySerialData = serial.Serial("com3", 9600, timeout=None)
inString = ""
multiplier = ""
numDataString = ""
multiplier2 = ""
numDataString2 = ""
data = []
data2 = []
graphCount = 0
i = datetime.now()

fileName = str(i.year) + "-" + str(i.month) + "-" + str(i.day) + "_" + str(i.hour) + "hrs_" + str(i.minute) + "mins_" + str(i.second) + "secs.txt"
f = open('RecordedData\\' + fileName, 'a')
f.write("Data from " + str(i.year) + "-" + str(i.month) + "-" + str(i.day) + " at " + str(i.hour) + ":" + str(i.minute) + ":" + str(i.second) + '\n')

def teensyData():
    global data
    global data2
    global graphCount
    global multiplier
    global multiplier2
    global inString
    global numDataString
    global numDataString2
    if (teensySerialData.inWaiting()>0):
        if (startBtnClicked == True):
            inString = ord(teensySerialData.read())
            if(multiplier == ""):
                multiplier = float(inString)
            elif(numDataString == ""):
                numDataString = float(inString)
                numData = multiplier*256 + numDataString
                numData = numData*3.3/1024
                data.append(numData)
                f.write("Data 1: " + str(numData) + '\n')
            elif(multiplier2 == ""):
                multiplier2 = float(inString)
            elif(numDataString2 == ""):
                numDataString2 = float(inString)
                numData = multiplier2*256 + numDataString2
                numData = numData*3.3/1024
                data2.append(numData)
                f.write("Data 2: " + str(numData) + '\n')
                multiplier = ""
                numDataString = ""
                multiplier2 = ""
                numDataString2 = ""
                graphCount = graphCount + 1
        global resetBtnClicked
        if (resetBtnClicked == True):
            data = []
            data2 = []
            resetBtnClicked = False
        #print(myData)
        
def update():
    global data
    global graphCount
    teensyData()
    if (graphCount == 3300):
        curve.setData(data)
        curve2.setData(data2)
        graphCount = 0
        #data = []
    
timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(0)

## Start the Qt event loop
app.exec_()
```
