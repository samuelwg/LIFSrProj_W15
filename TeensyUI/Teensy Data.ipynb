```python
from pyqtgraph import QtGui, QtCore
import pyqtgraph as pg
import numpy as np
import sys
import serial
from datetime import datetime
import time

## Always start by initializing Qt (only once per application)
app = QtGui.QApplication([])

## Define a top-level widget to hold everything
w = QtGui.QWidget()
w.resize(1000,600)
w.setWindowTitle('Voltage Plots')

resetBtnClicked = False
startBtnClicked = False

def startButtonClicked():
    global startBtnClicked
    global startBtn
    if (startBtnClicked == False):
        teensySerialData.flushInput()
        startBtnClicked = True
            
    elif (startBtnClicked == True):
        startBtnClicked = False
    
def resetButtonClicked():
    global resetBtnClicked
    resetBtnClicked = True
    
## Create widgets to be placed inside

startBtn = QtGui.QPushButton('Start')
startBtn.setToolTip('Click to begin graphing')
startBtn.resize(startBtn.sizeHint())
        
resetBtn = QtGui.QPushButton('Reset')
resetBtn.setToolTip('Click to clear graph')
resetBtn.resize(resetBtn.sizeHint())

startBtn.clicked.connect(startButtonClicked)
resetBtn.clicked.connect(resetButtonClicked)

xSamples = 100000
# Create plot widget
wGL = pg.PlotWidget()
wGL.setSizePolicy(QtGui.QSizePolicy.Expanding, QtGui.QSizePolicy.Expanding)
wGL.setYRange(0, 3.3)
wGL.setXRange(0, xSamples)
wGL.setLabel('top', text = "Plot 1")
sCurve = wGL.plot()

wGL2 = pg.PlotWidget()
wGL2.setSizePolicy(QtGui.QSizePolicy.Expanding, QtGui.QSizePolicy.Expanding)
wGL2.setYRange(0, 3.3)
wGL2.setXRange(0, xSamples)
wGL2.setLabel('top', text = "Plot 2")
pCurve = wGL2.plot()

## Create a grid layout to manage the widgets size and position
layout = QtGui.QGridLayout()
w.setLayout(layout)
#layout.setColumnStretch (1, 2)

## Add widgets to the layout in their proper positions
layout.addWidget(startBtn, 1, 0)   # start button
layout.addWidget(resetBtn, 2, 0)   #reset button
layout.addWidget(wGL, 1, 1, 1, 1)  # wGL goes on right side, spanning 3 rows
layout.addWidget(wGL2, 1, 2, 1, 1)

## Display the widget as a new window
w.show()

teensySerialData = serial.Serial("com32", 9600)
dataCount = 0
sData = np.zeros(xSamples)
pData = np.zeros(xSamples)
timeElapsed = 0
timeElapsedPrev = 0
graphCount = 0
i = datetime.now()

fileName = str(i.year) + "-" + str(i.month) + "-" + str(i.day) + "_" + str(i.hour) + "hrs_" + str(i.minute) + "mins_" + str(i.second) + "secs.txt"
f = open('RecordedData\\' + fileName, 'a')
f.write("#Data from " + str(i.year) + "-" + str(i.month) + "-" + str(i.day) + " at " + str(i.hour) + ":" + str(i.minute) + ":" + str(i.second) + '\n')
f.write("Timestamp,SiPM,PeakDetector\n")

def update():
    global xSamples
    global dataCount
    global sData
    global pData
    global graphCount
    global timeElapsed
    global timeElapsedPrev
    bufferSize = teensySerialData.inWaiting()
    #if(startBtnClicked == True):
        #print(bufferSize)
    runCount = bufferSize/8
    while (runCount > 0):
        if (startBtnClicked == True):
            inputBytes = teensySerialData.read(size = 8)
            timeByte3 = float(ord(inputBytes[0]))
            timeByte2 = float(ord(inputBytes[1]))
            timeByte1 = float(ord(inputBytes[2]))
            timeByte0 = float(ord(inputBytes[3]))
            sByte1 = float(ord(inputBytes[4]))
            sByte0 = float(ord(inputBytes[5]))
            pByte1 = float(ord(inputBytes[6]))
            pByte0 = float(ord(inputBytes[7]))
            timeElapsedPrev = timeElapsed
            timeElapsed = timeByte3*256*256*256 + timeByte2*256*256 + timeByte1*256 + timeByte0 #There are 8 bits in a byte, 2^8 = 256
            f.write(str(timeElapsed - timeElapsedPrev) + ",")
            if (timeElapsed - timeElapsedPrev > 150):
                print(str((timeElapsed-timeElapsedPrev)/100))
            numData = sByte1*256 + sByte0
            numData = numData*3.3/1024
            #s = np.roll(s, 1)
            #s[0] = numData
            sData[dataCount] = numData
            f.write(str(numData) + ",")
            numData = pByte1*256 + pByte0
            numData = numData*3.3/1024
            pData[dataCount] = numData
            f.write(str(numData) + '\n')
            graphCount = graphCount + 1
            dataCount = dataCount + 1
            if(dataCount >= xSamples):
                dataCount = 0
                sData = np.zeros(xSamples)
                pData = np.zeros(xSamples)
        global resetBtnClicked
        if (resetBtnClicked == True):
            s = np.zeros(xSamples)
            p = np.zeros(xSamples)
            resetBtnClicked = False
        if (graphCount >= 3000):
            sCurve.setData(sData[0:dataCount])
            pCurve.setData(pData[0:dataCount])
            graphCount = 0
            #data = []
        runCount = runCount - 1
    #print(teensySerialData.inWaiting())
    
timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(0)

## Start the Qt event loop
app.exec_()
```
